````python
# Define a function to determine if a collection is fiction, non-fiction, or undefined
def get_fiction_status(collection_name):
    if pd.isna(collection_name):
        return 'Undefined'
    collection_name_lower = str(collection_name).lower()
    if 'fiction' in collection_name_lower:
        return True
    elif 'nf' in collection_name_lower or 'nonfiction' in collection_name_lower:
        return False
    return 'Undefined'

# Apply this function to the 'Collection' column of the reference_data
reference_data['is_fiction'] = reference_data['Collection'].apply(get_fiction_status)

# Create a mapping dictionary from cleaned Wayfinder/Call Number Format to is_fiction status
# Use the same cleaning logic as for master_valid_categories to ensure consistency
def clean_for_mapping(s):
    if pd.isna(s):
        return None
    s = str(s)
    try:
        s = s.encode('latin1').decode('utf-8', errors='ignore')
    except (UnicodeEncodeError, UnicodeDecodeError):
        pass
    s = unidecode(s)
    s = s.replace(' /', '').replace('/', '').replace(';', '').replace('"', '').strip().upper()
    s = re.sub(r'\s*(\/|;|\\|\|)\s*', '', s)
    s = re.sub(r'^(JNF|JGN|DVD|AUDIO|BLU-RAY)\s*', '', s)
    s = re.sub(r'\s*\([0-9]+\)$', '', s)
    return s

wayfinder_fiction_map = {}
for index, row in reference_data.iterrows():
    wayfinder_cat = clean_for_mapping(row['Wayfinder'])
    if wayfinder_cat and wayfinder_cat not in wayfinder_fiction_map:
        wayfinder_fiction_map[wayfinder_cat] = row['is_fiction']

    call_num_format_cat = clean_for_mapping(row['Call Number Format'])
    if call_num_format_cat and call_num_format_cat not in wayfinder_fiction_map:
        wayfinder_fiction_map[call_num_format_cat] = row['is_fiction']


# Apply the mapping to the 'true_wayfinder' column in the main data DataFrame
data['fiction'] = data['true_wayfinder'].map(wayfinder_fiction_map)

# Fill any NaN values (where true_wayfinder did not match a key in the map) with 'Undefined'
data['fiction'] = data['fiction'].fillna('Undefined')

print("Distribution of the new 'fiction' column:")
display(data['fiction'].value_counts())

print("\nFirst 5 rows with 'fiction' column:")
display(data[['CALL NUMBER |a - WAYFINDER CATEGORY', 'true_wayfinder', 'fiction']].head())
````
